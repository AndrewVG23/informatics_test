```python
fname = '/Users/clare.huang/AskGene/data/external/lung_cancer_terminology.txt'
with open(fname, 'r') as f:
    lines = f.readlines()

with open(fname, 'r') as f:
    lines_copy = f.read()
```

```python
with open('rows_and_cols_terminology.txt', 'w') as f:
    for line in lines:
        title_check = re.search('[\d]{3} [\d]{4}/[\d] (.*) \n', line)
        subline_check = re.search('[\d]{4}/[\d] (.*) \n', line)
        if title_check:
            f.write(title_check.group(0))
            title_num = title_check.group(0)[:4]
            lines_copy = re.sub(title_check.group(0), '', lines_copy)
        elif subline_check:
            f.write(title_num + subline_check.group(0))
            lines_copy = re.sub(subline_check.group(0), '', lines_copy)        
        else:
            print(line)
```

```python
with open('big_topic.txt','w') as f:
    for line in lines_copy.split('\n'):
        if len(line) == 0:
            continue
        if 'Page' in line:
            continue
        elif 'LUNG & BRONCHUS C' in line:
            continue
        elif 'SEER SITE/HISTOLOGY VALIDATION LIST' in line:
            continue
        else:
            line = re.sub(r'http://seer.cancer.gov/icd-o-3/[\d]{3} of [\d]{3}', '', line)
        print(line)
        f.write(line + '\n')
```

```python
with open('rows_and_cols_terminology.txt', 'r') as f:
    load_cols = f.readlines()

make_df = list()
for line in load_cols:
    #print((line[:3], line[4:10], line[11:].rstrip()))
    make_df.append((line[:3], line[4:10], line[11:].rstrip()))

df = pd.DataFrame(make_df, columns=['Category_code', 'Sub-category code', 'Sub-category'])
```
