```python
#here we use a positive lookahead and a lazy quantifyer to find the first instance of "C[0-9]{3}" (e.g. C309)

def split_recode(text):
    for m in re.finditer(r'.*?(?=C[0-9]{3})',text):
        return text[:m.end()].strip()
```

```python
def split_recode_desc(text):
    for m in re.finditer(r'.*?(?=C[0-9]{3})',text):
        return text[m.end():].strip()
```
